```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("C:\\Users\\Micro\\Downloads\\gifts_by_occasion_1200.csv")

# 1. Number of gifts per occasion
occasion_counts = df['occasion'].value_counts()

plt.figure(figsize=(8,5))
occasion_counts.plot(kind='bar', color='skyblue')
plt.title("Number of Gifts per Occasion")
plt.xlabel("Occasion")
plt.ylabel("Gift Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Average price per occasion
avg_price = df.groupby("occasion")["price_usd"].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
avg_price.plot(kind='bar', color='salmon')
plt.title("Average Gift Price per Occasion")
plt.xlabel("Occasion")
plt.ylabel("Average Price (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Top 10 most common gift categories
top_categories = df['category'].value_counts().head(10)

plt.figure(figsize=(8,5))
top_categories.plot(kind='barh', color='lightgreen')
plt.title("Top 10 Gift Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.gca().invert_yaxis()
plt.show()
```

```python
# 4. Purchase channel distribution
channel_counts = df['purchase_channel'].value_counts()

plt.figure(figsize=(6,6))
channel_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#66c2a5','#fc8d62','#8da0cb'])
plt.title("Purchase Channel Distribution")
plt.ylabel("")
plt.show()
```

```python
# 5. Relationship vs Average Gift Price
rel_price = df.groupby("relationship")["price_usd"].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
rel_price.plot(kind='bar', color='purple')
plt.title("Average Gift Price by Relationship")
plt.xlabel("Relationship")
plt.ylabel("Average Price (USD)")
plt.xticks(rotation=45)
plt.show()
```
